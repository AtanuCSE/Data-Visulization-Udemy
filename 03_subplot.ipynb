Subplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4])
y = np.array([6,4,2,1])

plt.subplot(1,2,1) # first 1 means total row, 2 means total column, 1 means this plot number
plt.plot(x,y)
plt.title('January')

x = np.array([1,2,3,4])
y = np.array([2,3,2,8])

plt.subplot(1,2,2) # first 1 means total row, 2 means total column, 2 means this plot number
plt.plot(x,y)
plt.title('February')

plt.show()
```

```python
x = np.array([1,2,3,4])
y = np.array([6,4,2,1])

plt.subplot(2,1,1) # first 2 means total two rows, 1 means total one column, 1 means this plot number
plt.plot(x,y)


x = np.array([1,2,3,4])
y = np.array([2,3,2,8])

plt.subplot(2,1,2) # first 2 means total 2 rows, 1 means total one column, 2 means this plot number
plt.plot(x,y)

plt.show()
```
