# **Different Plots using matplotLib**
## Bar Plot, Histogram, Pie, Scatter, Box, 3D

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Bar Plot

```python
x = np.array(['A','B','C','D'])
y = np.array([5,8,2,10])

plt.bar(x,y, width=0.5)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([5,8,2,10])

plt.barh(x,y, color='r', height=0.2)
plt.show()
```

```python
Men = np.array([5,8,2,10, 5])
Women = np.array([4,7,1,3,4])

group = np.arange(5)

plt.grid()
plt.bar(group, Men, width=0.35, color='g')
plt.bar(group, Women, bottom=Men, width=0.5, color='b')

plt.xlabel('Participants')
plt.ylabel('Scores')

plt.xticks(group, ('P1','P2','P3','P4','P5'))
plt.yticks(np.arange(0, 21, 3))

plt.legend(['Men', 'Women'])
plt.show()
```

## Histogram

```python
x = np.random.normal(150, 10, 250)

plt.hist(x)
plt.show()
```

```python
x = np.random.normal(150, 10, 250)

plt.hist(x, density=True)
plt.show()
```

## Pie Chart

```python
x = np.array([3,7,10,5,2])

plt.pie(x)
plt.show()
```

```python
x = np.array([3,7,10,5,2])
lb = ['John', 'Mike', 'Claudia', 'Alice', 'Paris']

plt.pie(x, labels=lb)
plt.show()
```

```python
x = np.array([3,7,10,5,2])
lb = ['John', 'Mike', 'Claudia', 'Alice', 'Paris']
exp = [0, 0 ,0.2 ,0 ,0]

plt.pie(x, labels=lb, explode=exp, shadow=True, autopct='%1.0f%%')
plt.show()
```

## Scatter

```python
x1 = np.array([2,3,8,10, 12, 15, 20])
y1 = np.array([4,5,6, 9, 6, 12, 10])

plt.scatter(x1, y1)

x2 = np.array([3,1,7,7, 5, 10, 18])
y2 = np.array([6,7,8, 1, 9, 17, 8])

plt.scatter(x2, y2)


plt.show()
```

```python
x1 = np.array([2,3,8,10, 12, 15, 20])
y1 = np.array([4,5,6, 9, 6, 12, 10])

colors = [10, 20, 30, 40, 50, 60, 70]
size = [10, 23, 35, 47, 19, 41, 333]

plt.scatter(x1, y1, c=colors, cmap='viridis', s=size)
plt.colorbar()

plt.show()
```

## Box Plot

```python
np.random.seed(10)
data = np.random.normal(100, 10, 200)

fig = plt.figure(figsize=(10,7))

plt.boxplot(data)
plt.show()
```

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(200, 20, 400)
data3 = np.random.normal(150, 14, 200)
data4 = np.random.normal(300, 40, 500)
data = [data1, data2, data3, data4]

fig = plt.figure(figsize=(10,7))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data)
plt.show()
```

## 3D Plot

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')

x=[1,2,3,4,5]
y=[2,5,6,3, 8]
z = [10, 5, 20, 15, 10]

ax.plot3D(x,y,z, 'gray')
plt.show()
```
