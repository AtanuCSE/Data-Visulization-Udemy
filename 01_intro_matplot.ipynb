# **Matplotlib**

## Line

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.array([3,8,2,10])

plt.plot(y, linestyle='-.', color='#4c0', linewidth=5)
plt.show()
```

```python
y1 = np.array([3,8,2,10])
y2 = np.array([6,4,2,1])

plt.plot(y1, linestyle='-.', color='#4c0', linewidth=5)
plt.plot(y2, linestyle='-', color='r', linewidth=2)
plt.show()
```

## Marker

```python
y = np.array([3,8,2,10])
y2 = np.array([6,4,2,1])

plt.plot(y, linestyle='-', color='#4c0', linewidth=1, marker='*', ms=15, mec='r')
plt.plot(y2, linestyle='-', color='r', linewidth=2, marker='D', mfc='b')
plt.show()
```

## Fill Area

```python
n = 256

x = np.linspace(-np.pi, np.pi, n, endpoint=True)
y = np.sin(2*x)

plt.plot(x,y, color='blue', alpha=1.0)
plt.fill_between(x,y, 0, alpha=.1, color='blue')
plt.show()
```

## Legend and Label Manipulation

```python
x1 = np.array([2,3,8,10])
y1 = np.array([4,5,6, 9])

x2 = np.array([3,4,5,7])
y2 = np.array([6,8,5,7])

plt.plot(x1, y1, x2, y2, marker='o')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.legend(labels=('Company1', 'Company2'), loc='lower right')
plt.show()
```

## Gridlines

```python
x1 = np.array([2,3,8,10])
y1 = np.array([4,5,6, 9])

x2 = np.array([3,4,5,7])
y2 = np.array([6,8,5,7])

plt.plot(x1, y1, 'ys-')
plt.plot(x2, y2, 'go--')
plt.grid(True, axis='both', color='b', lw=0.25)
plt.xlabel('Month')
plt.ylabel('Profit')
plt.legend(labels=('Company1', 'Company2'), loc='lower right')
plt.show()
```
